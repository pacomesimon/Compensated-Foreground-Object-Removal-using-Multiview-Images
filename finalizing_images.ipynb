```python
def upsample_image(idx):
  if os.path.exists(f"{global_config.MODEL_MASKS_IMGS_PATH}{idx}.jpg") and os.path.exists(f"{global_config.INFERENCE_DATA_PATH}{idx}.jpg"):

    mask = cv2.imread(f"{global_config.MODEL_MASKS_IMGS_PATH}{idx}.jpg")
    # mask = cv2.erode(mask, np.ones((3, 3), np.uint8), iterations=1)
    # mask = cv2.dilate(mask, np.ones((3, 3), np.uint8), iterations=1)
    image1 = cv2.imread(f"{global_config.INFERENCE_DATA_PATH}{idx}.jpg")
    image2 = cv2.imread(f"{global_config.STITCHED_IMGS_PATH}{idx}.jpg")
    image1 = cv2.resize(image1, (image2.shape[1], image2.shape[0]))
    mask = cv2.resize(mask, (image2.shape[1], image2.shape[0]))
    # mask = cv2.erode(mask, np.ones((5, 5), np.uint8), iterations=1)
    # mask = cv2.dilate(mask, np.ones((3, 3), np.uint8), iterations=1)
    new_one = image1*((~mask)//255) + image2*((mask)//255)
  else:
    new_one = cv2.imread(f"{global_config.STITCHED_IMGS_PATH}{idx}.jpg")
  cv2.imwrite(f"{global_config.FINAL_OUTPUTS_PATH}{idx}.jpg", new_one)

# upsample_image(15)
for idx in range(20):
  upsample_image(idx)
```
